```python
import pandas as pd
```

```python
sub = pd.read_csv('152_029_xgboost.csv')
sub
```

```python
pd.set_option('display.max_rows', 500)
```

```python
sub['pred_new'] = sub['prediction']*0.7 + sub['prediction'].shift(1)*0.15 + sub['prediction'].shift(-1)*0.15 
display(sub.head(100))
sub['prediction'] = sub.apply(lambda x: x['pred_new'] if x['prediction'] != 0 else 0, axis=1)
```

```python
sub['prediction'].to_csv('xgboost_smooth.csv', index=True)
```
